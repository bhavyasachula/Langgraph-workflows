```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class BatsmenState(TypedDict):
    runs:int
    balls:int 
    fours:int
    sixes:int

    strikeRate:float
    ballsperboundary:float
    boundarypercent:float
    summary:str
```

```python
def calculateSR(state:BatsmenState):
    sr =(state['runs']/state['balls'])*100
    state['strikeRater'] = sr
    return {'strikeRate': sr}
```

```python
def calculateBallsPerBoundary(state:BatsmenState):
    bpb = state["balls"]/(state["fours"]+state["sixes"])
    state['ballsperboundary'] = bpb
    return {'ballsperboundary':bpb}
```

```python
def calculateBoundaryPercent(state):
    boundarypercent =( ((state["fours"]*4) + (state["sixes"]*6))/state["runs"])*100
    state["boundarypercent"]= boundarypercent
    return {"boundarypercent" : boundarypercent}
```

```python
def summary(state:BatsmenState):
    summary=f"""
        Strike Rate - {state["strikeRate"]}
        Balls per Boundary - {state["ballsperboundary"]}
        Boundary Percentage - {state["boundarypercent"]}
    """
    return {"summary":summary}
```

```python
graph = StateGraph(BatsmenState)
```

```python
graph.add_node("sr",calculateSR)
graph.add_node("bpb",calculateBallsPerBoundary)
graph.add_node("boundaryper",calculateBoundaryPercent)
graph.add_node("summary",summary)

graph.add_edge(START,"sr")
graph.add_edge(START,"bpb")
graph.add_edge(START,"boundaryper")
graph.add_edge("sr","summary")
graph.add_edge("bpb","summary")
graph.add_edge("boundaryper","summary")
graph.add_edge("summary",END)
```

```python
workflow = graph.compile()
```

```python
intial_state={
    "runs":100,
    "balls":50,
    "fours":6,
    "sixes":4
}
finalstate=workflow.invoke(intial_state)
print(finalstate)
```
