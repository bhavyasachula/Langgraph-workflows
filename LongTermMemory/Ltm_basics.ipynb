```python
from langgraph.store.memory import InMemoryStore
```

```python
store = InMemoryStore()
```

```python
namespace1 = ("users","u1") # namespace is like a folder structure where we pass the folder like users has u1 
```

__STORING USER DATA__ 

```python
# Create memories 
store.put(namespace1,"1",{"data":"user likes pizza"})
store.put(namespace1,"2",{"data","user likes burger"})
```

```python
namespace2 = ("users","u2") # to create the namespace u must use tuple with string in it to seperate use ","
```

```python
store.put(namespace2,"1",{"data":"user like to code in python"})
store.put(namespace2,"1",{"data":"user like pasta"})
```

```python
store.get(namespace1,"1")
```

```python
#retrive all memories
items = store.search(namespace1)
for i in items:
    print(i.value)
```
