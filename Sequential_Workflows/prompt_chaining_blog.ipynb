```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
from langchain_groq import ChatGroq
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os 
```

```python
class blogState(TypedDict):
    title:str
    outline:str
    content:str
```

```python
llm = ChatOpenAI(model="openai/gpt-oss-120b")
graph = StateGraph(blogState)
```

```python

def gen_outline(state:blogState)-> blogState:
   #extract the title
    title = state['title']
    

    #call llm outline
    prompt1 = f"generate me the outline on the basis of this title:{title}"
    response = llm.invoke(prompt1).content
    #update state
    state['outline'] = response
    #return state
    return state
```

```python
def gen_blog(state:blogState)->blogState:
    #generate blog based on the title and the outline 
    title = state['title']
    outline = state['outline']

    #prompt to generate the blog
    prompt2=f"generate the blog based on the title :{title} and outline : {outline}" 

    #call the llm to generate blog
    response = llm.invoke(prompt2).content

    #update the state
    state['content']= response
   

    return state
```

```python
graph.add_node("gen_outline",gen_outline)
graph.add_node("gen_blog",gen_blog)

graph.add_edge(START,"gen_outline")
graph.add_edge("gen_outline","gen_blog")
graph.add_edge("gen_blog",END)
```

```python
workflow = graph.compile()
```

```python
intial_state = {'title':'Orchestration in Ai'}
final_state = workflow.invoke(intial_state)
print(final_state['outline'])
```
