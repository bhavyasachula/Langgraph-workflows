```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
from langchain_groq import ChatGroq
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os 
```

```python
load_dotenv()
```

```python
class llmState(TypedDict):
    question:str
    answer:str
```

```python
llm = ChatOpenAI(model="openai/gpt-oss-120b")
llm
```

```python
def llm_q_a(state : llmState)->llmState:
    #extract question from the state
    question= state['question']

    #add question to prompt
    prompt = f"answer the following question:{question}"

    #invoke the llm and give the prompt 
    response = llm.invoke(prompt)

    #store the response to the answer state
    state['answer']=response.content

    return state
```

```python
graph = StateGraph(llmState)
```

```python
#added the node to graph 
graph.add_node("Q&A",llm_q_a)

#connecting  the edges
graph.add_edge(START,"Q&A")
graph.add_edge("Q&A",END)

#compile the graph
workflow=graph.compile()
```

```python
intial_state={"question":"wht is orchestration in Ai"}
final_state = workflow.invoke(intial_state)
print(f"Ai answer :{final_state['answer']}")
```
